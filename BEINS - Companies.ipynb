```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Data_Companies.csv")
```

```python
data.head()
```

```python
data['DMA_CHANGE'] = ((data['DMA 50'] - data['DMA 200'])/data['DMA 200'])*100

list = []
for i in abs(data['DMA_CHANGE']):
    if i < 2:
        list.append('Conservative')
    elif i < 5:
        list.append('Moderatively Conservative')  
    elif i < 10:
        list.append('Moderatively Aggresive')    
    elif i < 15:
        list.append('Aggresive')   
    else:
        list.append('Very Aggresive')    
          
data['Risk Profile'] = list      
```

# Mapping Customer to companies

```python
data['PE_DIFF'] = data['Industry PE']-data['Price to Earning']
data['PRICE_VAR_HIGH'] = (data['Current Price']/data['High price all time'])*100
data['PRICE_VAR_LOW'] = (data['Low price all time']/data['Current Price'])*100
```

```python
def Score_cap1(data,col):
    list = []
    for i in data[col]:
        if i > data[col].quantile(0.90):
            list.append(2)
        elif i > data[col].quantile(0.75):
            list.append(1)
        else:
            list.append(0)
     
    data[col+'_BUCKET'] = list
    
    
def Score_cap2(data,col):
    list = []
    for i in data[col]:
        if i < data[col].quantile(0.10):
            list.append(2)
        elif i < data[col].quantile(0.25):
            list.append(1)
        else:
            list.append(0)
     
    data[col+'_BUCKET'] = list    
```

```python
list_upper = ['PE_DIFF','Return on equity','Return on assets','Return on capital employed','Piotroski score',
              'Dividend yield','Market Capitalization','Promoter holding','FII holding','YOY Quarterly profit growth',
              'PRICE_VAR_HIGH','OPM']

list_lower = ['Industry PE','Price to book value','Debt to equity','PRICE_VAR_LOW','Pledged percentage']
```

```python
for i in list_upper:
    Score_cap1(data,i)
    
for i in list_lower:
    Score_cap2(data,i)    
```

```python
data['Company_score'] = (data['PE_DIFF_BUCKET']+data['Return on equity_BUCKET']+data['Return on assets_BUCKET']+data['Return on capital employed_BUCKET']+data['Piotroski score_BUCKET']+data['Dividend yield_BUCKET']+data['Market Capitalization_BUCKET']+data['Promoter holding_BUCKET']+data['FII holding_BUCKET']+data['YOY Quarterly profit growth_BUCKET']+data['PRICE_VAR_HIGH_BUCKET']+data['OPM_BUCKET'] + data['Industry PE_BUCKET']+data['Price to book value_BUCKET']+data['Debt to equity_BUCKET']+data['PRICE_VAR_LOW_BUCKET']+data['Pledged percentage_BUCKET'])/34
data['Company_Rank'] = data.groupby(['Risk Profile'])['Company_score'].rank(method = 'first', ascending = False)
```

```python
data.to_csv('Data_Companies.csv',index = False)
```
